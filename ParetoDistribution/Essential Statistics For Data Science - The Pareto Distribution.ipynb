#### Statistics IS NOT Intuitive

Lack of domain knowledge can often be managed with the help of domain experts. Lack of programming skills can be managed with the help of engineers. Knowledge of statistics on the other hand is a critical skill: It's very easy to interpret statistical results incorrectly. Here's an example:

The following are admission figures from the University of California, Berkeley for the fall of 1973:

|Applicants|Total Admitted|Percentage Admitted|
|----------|--------------|-------------------|
|Men	   |8442	      |44%                |
|Women	   |4321	      |35%                |

They showed that men applying were more likely than women to be admitted, and the large difference makes unlikely that it's due to chance. Gender bias?

When examining individual departments, it appeared that the majority of the departments were significantly more biased against men rather:

|Department|Men Total|Men Percentage|Women Total|Women Percentage|
|----------|---------|--------------|-----------|----------------|
|A         |825      |62%           |108        |82%             |
|B         |560      |63%           |25         |68%             |
|C         |325      |37%           |593        |34%             |
|D         |417      |33%           |375        |35%             |
|E         |191      |28%           |393        |24%             |
|F         |373      |6%            |341        |7%              |

Gender bias against men then? Turned out not to be the case: This [research paper by Bickel et. all](http://homepage.stat.uiowa.edu/~mbognar/1030/Bickel-Berkeley.pdf) concluded that majority of women applied to more competitive departments with low rates of admission, whereas the majority of men applied to less-competitive departments with high rates of admission.

This type of statistical phenomena, where a trend appears in groups of data but disappears (or reverses) when these groups are combined is a common pattern in data and it's named ["The Simpson's Paradox"](https://en.wikipedia.org/wiki/Simpson%27s_paradox).

Statistics IS NOT intuitive at all! It needs to be learned. You won't be able to wing it...

## The Pareto Distribution

A data scientist needs to know at least the most common statistics distributions. I'm starting with one which I believe is fairly neglected in statistics courses. I know that I haven't seen it in ANY of the statistics and data analytics courses I took.

As a matter of fact, I was introduced to Pareto distributions by Tim Ferriss' book on productivity and time management: "The 4-hour Workweek" (or more specfically to the "80-20 rule" or "Pareto principle", which we will discuss later on).

Further research showed me that this distribution is indeed more widely known in business and management circles.

That's no coincidence, given that the Pareto distribution was originally applied to describe the distribution of wealth in a society. In fact, it also applies to many phenomena in the fields of economics and productivity.

Note that, eventually the simulation reaches a point of equilibrium, with the distribution roughly following the shape of a power-law function.

That might be counter-intuitive for you, depending on your ideology. We start from a state of "fairness" (all agents start with the same amount of money) and end up in a state of "unfairness" (the majority of money concentrated in the hands of a few agents). Note also that these agents are not malicious savy businessman: They execute transactions at random and still the point of equilibrium isn't an state where the money is uniformly distributed! They have exactly the same "negotiation skills", nothing special about any of them!


## How a Pareto Distribution Looks Like?

### The Power Function

This distribution follows a "power-law", which simply means that the function's output is a power of the input:

$F(x) = b \cdot x^a$

where:

* $b$ is a constant (just a number).
* $a$ is a constant (just a number).
* x is the input.
* F(x) is the output.

Here's the power-law function as a Python function:

````python
import numpy as np

def f(x, b, a):
    return b * (x ** a)

x = np.linspace(1.0, 200.0, 10000, endpoint=False)
y = [f(x_i, b=1.0, a=-2.0) for x_i in x]
````

We are setting "x" with 10,000 data points uniformily distributed in the interval interval [1.0, 200.0]. "y" is generated by applying the power-law function "f" over each data point in "x".

But a picture is worth more than a 1,000 words, so let's plot the data points:

````python
import matplotlib.pyplot as plt

def set_plot_ticks(max_x, max_y, x_step, y_step):
    if x_step is not None: plt.xticks(np.arange(0, max_x + x_step, step=x_step))
    if y_step is not None: plt.yticks(np.arange(0, max_y + y_step, step=y_step))
    
def set_plot_axis(max_x, max_y, x_step, y_step):
    plt.ylabel('F(X)')
    plt.xlabel('X')
    plt.xlim(0, max_x)
    plt.ylim(0, max_y)
    set_plot_ticks(max_x, max_y, x_step, y_step)

def line_plot(x, y, max_x = None, x_step = None, max_y = None, y_step = None, figsize=(10, 10)):
    if max_x is None: max_x = max(x)
    if max_y is None: max_y = max(y)
    
    figure = plt.figure(figsize=figsize)
    plt.plot(x, y)
    set_plot_axis(max_x, max_y, x_step, y_step)

    return figure
    
figure = line_plot(x, y, max_x = 20, x_step = 1, y_step=0.1)
figure.savefig('power-decay-function1', bbox_inches='tight')
plt.show()
````

Again, don't mind so much about the ploting code at this moment.

Note that $a$ could be positive. Let's make $a = 2.0$ and plot the data points:

````python
y = [f(x_i, b=1.0, a=2.0) for x_i in x]

figure = line_plot(x, y)
figure.savefig('power-growth-function1')
plt.show()
````

### The Pareto Probablity Density Function


We get the probability density function for a Pareto distribution by setting specific values of "a" and "b". The power-law function for a Pareto distribution is the following:

$F(x) = (\alpha \cdot {x_m}^\alpha) \cdot {x^{-(\alpha + 1)}}$

Note that this is not different from our generic power-law function: The Pareto probability density function is only a special case where $b = \alpha \cdot {m_x}^\alpha$ and $a = -(\alpha + 1)$.

The parameters $\alpha$ and $m_x$ represent the following following properties of the Pareto distribution:

* $x_m$ is the lower bound of the data (the lowest value of the input $x$).
* $\alpha$ is the shape parameter (how fast the function decays or grows).

Let's create a visualization of the Pareto probability density function. The following function not only plots the data points, but also computes the area under the curve:

````python
from scipy.integrate import simps

def line_plot_with_area(x, y, max_x=None, x_step = None, max_y=None, y_step = None, figsize=(10, 10)):
    figure = line_plot(x, y, max_x, x_step, max_y, y_step, figsize)
    if max_y is None: max_y = max(y)
    area = np.round(simps(y, x), decimals=2)
    plt.fill_between(x, 0, y)
    plt.text(0.5 * (min(x) + max_x), 0.5 * (min(y) + max_y),
             "area = %.1f" % area, horizontalalignment='center', fontsize=20)
    return figure
````

Let's build a power-law function with $x_m = 1$ and $\alpha = 1$, generate and plot the data points:

````python
x_m = 1
alpha = 1

y = [f(x_i, b=(alpha * (x_m ** alpha)), a=-(alpha + 1)) for x_i in x]
figure = line_plot_with_area(x, y, max_x = 20, x_step = 1, y_step=0.1)
figure.savefig('pareto-pdf1')
plt.show()
````

````python
def f(x, b, a):
    return b * (x ** a)


x = np.linspace(1.0, 200.0, 10000, endpoint=False)
y = [f(x_i, b=(alpha * (x_m ** alpha)), a=-(alpha + 1)) for x_i in x]
````

Note that the function is defined from $x = 1$ to infinity. In our example, $x$ goes from 1.0 to 200.0 though, so the area we are calculating is an approximation. Still, given that the function decreases fast, we get pretty darn close to one when calculating the area under the curve from $x = 1.0$ to $x = 200.0$ only!

An important fact is that this function describes a probability density, thus function parameters have been defined in such a way that the area under the curve is equal to 1.0.

## The Pareto Principle

The Pareto principle simply states the following:

> “~80% of the effects come from ~20% of the causes.

Pareto first noticed that about 80% of Italy's land was owned by 20% of the population. But the principle holds in a global scale: The 1992 United Nations Development Program Report showed that distribution of global income is very uneven, with the richest 20% of the world's population controlling 82.7% of the world's income.

### Examples of the Pareto Principle

The principle does not only apply to economics though. Here are some other real-life examples:

* Microsoft noted that by fixing the top 20% of the most-reported bugs, 80% of the related errors and crashes in a given software would be eliminated.

* In load testing for web sites is common practice to estimate that 80% of the traffic will occur during a particular 20% of the total time period.

* In baseball to determine a player's overall importance to a team: "15% of the all the players last year produced 85% of the total wins with the other 85% of the players creating 15% of the wins.

* The Dunedin Study has found that 80% of crimes are committed by 20% of criminals.

### Another Example, The Causes of Holodomor

Here's an example of the Pareto Principle which I thought to be deserving of its own section.

You might be acquainted with [Holomor](https://en.wikipedia.org/wiki/Causes_of_the_Holodomor) from history class. Holomor was a famine that ravaged the Soviet Ukraine in the 1930's. In his efforts to reshape Soviet society and eliminate inequality, [Joseph Stalin](https://en.wikipedia.org/wiki/Joseph_Stalin) ordered arrests, deportations, and executions of better-off peasants and their families. They were labeled "Kulaks" and declared enemies of the state.

Unfortunately, the Soviet state wasn't aware of the Pareto distribution, which not only applies to the distribution of wealth, but also to the distribution of productivity: These "richer peasants", the Kulaks, not only concentrated wealth, but productivity as well. The Kulaks were responsible for over 80% of the food production. Their elimination caused this terrible tragedy!

### From Where the Pareto Principle Comes From?

The Pareto principle, also known as the 80-20 rule comes from assuming a particular value for the shape parameter:

$\alpha = log4(5) = \frac{log(5)}{log(4)}$, which is approximately equal to $1.161$.

Let's generate and plot the data points for this particular case:

````python
%matplotlib inline

alpha = np.log(5) / np.log(4)

y = [f(x_i, b=(alpha * (x_m ** alpha)), a=-(alpha + 1)) for x_i in x]
figure = line_plot_with_area(x, y, max_x = 20, x_step = 1, y_step=0.1)
figure.savefig('pareto-principle-pdf1')
plt.show()
````

Not much of a difference from our previous plot. In the next plot we will make the difference more evident.

The Pareto principle states that 80% of the output can be explained by 20% of the input. Regarding specifically the probability density function, that means that 20% of interval explains 80% of the total probability (that is, the area under the curve in the interval is equal to 0.8). The remaining 80% of the interval therefore explains the remaining 20% percent of the probability (the area under the curve in the interval is equal to 0.2).

Around $x = 16$, the function becomes very small, so let's take the interval from $x = 1$ to $x = 16$ (length equal to 15) to be 100% of the interval. That's somewhat arbritary so the calculations become easier.


This makes the interval from $x = 1$ to $x = 4$ (length equal to 3, which is equal to 20% of 15) our 20% interval. If we calculate the area under the curve:

````python
x = np.linspace(start=1.0, stop=4, num=100000, endpoint=False)
y = [f(x_i, b=(alpha * (x_m ** alpha)), a=-(alpha + 1)) for x_i in x]

figure = line_plot_with_area(x, y, max_x = 20, x_step = 1, y_step=0.1)
figure.savefig('pareto-principle-pdf-20-percent-interval-1')
plt.show()
````

20% of the interval accounts for 0.8 or 80% of the total probability. The remaining of the interval correspond to the remaining 80%. If we calculate the area under the curve:

````python
x = np.linspace(start=4, stop=16, num=100000, endpoint=False)
y = [f(x_i, b=(alpha * (x_m ** alpha)), a=-(alpha + 1)) for x_i in x]

figure = line_plot_with_area(x, y, max_x=20, max_y=1, x_step = 1, y_step=0.1)
figure.savefig('pareto-principle-pdf-80-percent-interval-1')
plt.show()
````

The remaining 80% of the interval accounts for only 0.2 or 20% of the total probability!

Would be nice to see the two areas in a single plot. The following function breaks the interval in two and plot their correspondent areas:

````python
def plot_area(x, y, min_x, max_x, min_y, max_y, area, text_position, color):
    plt.fill_between(x, 0, y)
    plt.text(text_position * (min_x + max_x), 0.5 * (min_y + max_y),
             "area = %.1f" % area, horizontalalignment='center', fontsize=20, color=color)  

def line_plot_with_split_area(x, y, x_split, max_x=None, x_step = None, max_y=None, y_step = None):
    figure = line_plot(x, y, max_x, x_step, max_y, y_step)
    if max_y is None: max_y = max(y)
    min_x = min(x)
    min_y = min(y)

    split_index = np.where(x == x_split)[0] + 1
    x1, x2 = np.split(x, split_index)
    y1, y2 = np.split(y, split_index)
    area1 = np.round(simps(y1, x1), decimals=2)
    area2 = np.round(simps(y2, x2), decimals=2)
    
    plot_area(x1, y1, min_x, max_x, min_y, max_y, area1, 0.25, 'blue')
    plot_area(x2, y2, min_x, max_x, min_y, max_y, area2, 0.75, 'red')
    
    return figure
````

Let's put the two areas together in a single plot covering the whole interval:

````python
x = np.linspace(start=1, stop=16, num=300000, endpoint=False)
y = [f(x_i, b=(alpha * (x_m ** alpha)), a=-(alpha + 1)) for x_i in x]

figure = line_plot_with_split_area(x, y, x_split = 4, max_x = 20, x_step = 1, y_step=0.1)
figure.savefig('pareto-principle-pdf-20-and-80-percent-interval-1')
plt.show()
````

To do the Pareto analysis we need to group the complaints by issues and count the number of issues of each type.

We should also compute the cummulative percentage contribution of each issue type from the larger to the smaller contributor:

````python
def column_data(data, column_name):
    return data[[column_name]]

def group_by_column_and_count(column_data, column_name):
    return column_data.groupby([column_name]).size().reset_index(name="count")

def sort_by_count(column_counts):
    return column_counts.sort_values(by=["count"], ascending=False).reset_index()

def cumulative_percentage(column_counts):
    total_count = column_counts["count"].sum()
    return 100 * column_counts["count"].cumsum() / total_count

def get_pareto_stats(data, column_name):
    stats = column_data(data, column_name)
    stats = group_by_column_and_count(stats, column_name)
    stats = sort_by_count(stats)
    stats["cummulative_percentage"] = cumulative_percentage(stats)
    return stats

pareto_stats = get_pareto_stats(mortage_complaints, 'Issue')
pareto_stats
````

The function groups the records by the value of the "Issue" column and computes the count for each type of "Issue".

It also computes the cummulative percentage of the issues covered by each type of "Issue" from the largest to the smallest contributor.

From the output, you may notice that the first two issue types are roughly responsible for 80% of the total complaints! There it is the Pareto principle for everyone to see!

But it's much easier to visualize this in a plot:

````python
%matplotlib inline

import matplotlib.pyplot as plt

def set_count_axis(count_axis, stats, column_name):
    count_axis.set_xticks(stats.index.values)
    count_axis.set_xticklabels(stats[column_name].values, rotation=90, va='bottom')
    count_axis.tick_params(axis="x",direction="in", pad=-10)
    count_axis.set_ylabel("Frequency", size=16) 
    
def set_cumulative_percentage_axis(cumulative_percentage_axis):
    cumulative_percentage_axis.set_yticks(np.arange(0., 101., 5.))
    cumulative_percentage_axis.set_ylabel('Cumulative Percentage', size=16)
    
def create_bar_plot(count_axis, stats, column_name):
    count_axis.bar(stats.index.values, stats["count"].values, alpha = 0.5, color='orange')
    set_count_axis(count_axis, stats, column_name)    
    

def create_line_plot(cumulative_percentage_axis, stats):
    cumulative_percentage_axis.plot(stats.index.values, stats.cummulative_percentage.values, linestyle="-.")
    set_cumulative_percentage_axis(cumulative_percentage_axis)    
    
def pareto_chart(data, column_name):
    stats = get_pareto_stats(data, column_name)
    
    figure, count_axis = plt.subplots(figsize=(16, 8))
    create_bar_plot(count_axis, stats, column_name)
    
    cumulative_percentage_axis = count_axis.twinx()
    create_line_plot(cumulative_percentage_axis, stats)

    plt.title('Pareto Chart', fontweight="bold", fontsize=20, y=1.1)
    plt.grid()
    return figure

figure = pareto_chart(mortage_complaints, 'Issue')
figure.savefig('pareto-chart-1')
plt.show()
````

Note that the first type of "Issue" is responsible for roughly 80% of all the issues. The cummulative percentage (dotted line plot) reaches rougly 80% with the contributions of the two first bars.

The code in this plot is reasonably general, so you may provide a different data frame and specify a different column name to have a Pareto chart built for it:

```
pareto_chart(<my data frame>, <my categorical column>)
```
